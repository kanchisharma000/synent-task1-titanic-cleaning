### Objective
Clean and prepare the Titanic dataset for analysis by handling missing values, removing duplicates, converting data types, and renaming columns.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
df.head() 
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

## Data Cleaning

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df = df.drop(columns=['Cabin'])
```

```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.to_csv("titanic_cleaned.csv", index=False)
```

## Conclusion

- Missing values handled using median and mode
- Cabin column removed due to excessive missing values
- Duplicate records removed
- Column names standardized
- Dataset is now clean and ready for analysis
